```python
#!/usr/bin/env python3
"""Peek 一个样本 JSON 的字段与取值。

用法：
    python peek_train_json.py <json_file_path> [index]

- 顶层是 list 时取 data[index]，默认 index=0；顶层是 dict 时直接展平。
- 展平规则：
  * dict → path 用 `.` 拼接子键；
  * list → path 用 `.` 拼接下标（0/1/2...）；
  * 叶子（非 dict/非 list，或空 dict/空 list）触发一次输出。
- 每个叶子按下面格式打印：
    ========== <path> ==========
    <value>
"""
import json
import sys
from pathlib import Path


def flatten(obj, path=""):
    if isinstance(obj, dict):
        if not obj:
            yield path or "<root>", {}
            return
        for k, v in obj.items():
            sub = f"{path}.{k}" if path else str(k)
            yield from flatten(v, sub)
    elif isinstance(obj, list):
        if not obj:
            yield path or "<root>", []
            return
        for i, v in enumerate(obj):
            sub = f"{path}.{i}" if path else str(i)
            yield from flatten(v, sub)
    else:
        yield path or "<root>", obj


# if len(sys.argv) not in (2, 3):
#     print("usage: python peek_train_json.py <json_file_path> [index]", file=sys.stderr)
#     sys.exit(2)
# p = Path(sys.argv[1])
# idx = int(sys.argv[2]) if len(sys.argv) == 3 else 0

p = Path('/home/work/slamm/youwei.wang/lora_sft_demo/train_data/train_qwen35vl_9b_extract_attrs_tags_h100_8gpu.json')

idx = 13

with p.open("r", encoding="utf-8") as f:
    data = json.load(f)

if isinstance(data, list):
    if not data:
        print("[warn] empty list", file=sys.stderr)
        sys.exit(0)
    sample = data[idx]
    print(f"# file={p}  total_samples={len(data)}  index={idx}")
else:
    sample = data
    print(f"# file={p}  root=dict")

for path, value in flatten(sample):
    print(f"========== {path} ==========")
    print(value)

```
